```python
#prtn = []
def intact2rel(intact,uprot_dict_human,uprot_dict_sars,itmpGraph):
    for i in range(len(intact)):
        a = intact['# ID(s) interactor A'][i].split(':')[1]
        a_copy = a
        #print(a_copy)
        
        if a in uprot_dict_human.keys():
            p1 = Protein(namespace='HGNC',name=a)
            #print('human',p1)
            if uprot_dict_human[a]['Gene'] != {}:
                a = uprot_dict_human[a]['Gene']
                p1=Protein(namespace='HGNC',name=a)
                #print('human2',p1)

        if a in uprot_dict_sars.keys():
            p1 = Protein(namespace='VP',name=a)
            
            if uprot_dict_sars[a]['Gene'] != {}:
                a = uprot_dict_sars[a]['Gene']
                p1=Protein(namespace='VP',name=a)
                
        else:
            p1=Protein(namespace='HGNC',name=a)
            
        #break

        b = intact['ID(s) interactor B'][i].split(':')[1]
        #print('b',b)
        b_copy = b
        #print(b_copy)
        
        if b in uprot_dict_human.keys():
            p2 = Protein(namespace='HGNC',name=b)
            #print('b',p2)
            if uprot_dict_human[b]['Gene'] != {}:
                b = uprot_dict_human[b]['Gene']
                p2=Protein(namespace='HGNC',name=b)
                #print('b2',p2)

        if b in uprot_dict_sars.keys():
            #print('sars',b)
            p2 = Protein(namespace='VP',name=b)
            if uprot_dict_sars[b]['Gene'] != {}:
                b = uprot_dict_sars[b]['Gene']
                p2=Protein(namespace='VP',name=b)
        
        else:
            p2=Protein(namespace='HGNC',name=b)
        
        dbio = 'https://3dbionotes.cnb.csic.es/?queryId='
        evi = intact['Interaction identifier(s)'][i].split("|")[0].split(":")[1]
        itmpGraph.add_association(p1,
                                  p2,
                                  citation = 'IntAct',
                                  evidence = 'https://www.ebi.ac.uk/intact/details/interaction/'+evi)
        
#         print(p1)
#         print(p2)
#         #break
        
        #create links to 3Dbionotes
        
        nx.set_node_attributes(itmpGraph,{p1:'https://3dbionotes.cnb.csic.es/?queryId='+a_copy},'3Dbio')
        nx.set_node_attributes(itmpGraph,{p2:'https://3dbionotes.cnb.csic.es/?queryId='+b_copy},'3Dbio')
        
        
        
        
    return(itmpGraph)
    
    
```

```python
## to Neo4j
# Exporting graph to Neo4J
import py2neo
node_map = {}

NEO4J_USER = 'yojana'
NEO4J_PASS = 'abc'

neo_connection = py2neo.Graph('bolt://localhost:7687', auth=('neo4j', 'itmp'),name = 'intact')
tx = neo_connection.begin()
neo_connection.delete_all()

nodes = list(intact_graph)
#print(nodes)

dbio = nx.get_node_attributes(intact_graph,'3Dbio')

for node in tqdm(nodes, desc="nodes"):
    attrs = {"namespace": node.namespace}

    if node.name and node.identifier:
        attrs["name"] = node.name
        attrs["identifier"] = node.identifier
    elif node.identifier and not node.name:
        attrs["name"] = node.identifier
    elif node.name and not node.identifier:
        attrs["name"] = node.name

    if node in dbio:
        attrs['3Dbio']=dbio[node]
    node_map[node] = py2neo.Node(node.function, **attrs)

    tx.create(node_map[node])

edges = intact_graph.edges(keys=True, data=True)

for u, v, key, node in tqdm(edges, desc="edges"):
    rel_type = node['relation']

    d = node.copy()
    del d['relation']

    attrs = {}

    annotations = d.pop('annotations', None)
    if annotations:
        for annotation, values in annotations.items():
            attrs[annotation] = list(values)

    citation = d.pop('citation', None)
    if citation:
        attrs['citation'] = citation.curie

    if 'evidence' in d:
        attrs['evidence'] = d['evidence']

    rel = py2neo.Relationship(node_map[u], rel_type, node_map[v], key=key, **attrs)
    tx.create(rel)
    
tx.commit()   
```
